```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf1 = adfuller(df['Adidas'], regression='ct', autolag='AIC', regresults=True)
adf1[3].resols.summary()
```

```python
# attention ici √† bien changer l'option "regression" et mettre "c" au lieu de "ct"
adf1 = adfuller(df['Adidas'], regression='c', autolag='AIC', regresults=True)
adf1[3].resols.summary()
```

```python
# attention ici √† bien changer l'option "regression" et mettre "c" au lieu de "ct"
adf1 = adfuller(df['Adidas'], regression='n', autolag='AIC', regresults=True)
adf1
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
```

```python
l_explicatives =  ['rAdidas','rBayer'] 
for n,var in enumerate(l_explicatives):
    print('le VIF de la variable', var, ' est : ',VIF(df[l_explicatives].dropna().values,n))
```

```python
import numpy as np
from numpy.linalg import det
```

```python
D = det(df[l_explicatives].corr())

#(1/2 k(k+1)) 
    # calcul de la stat : T=1261 et k=2
FG = -(1261 - (1/6) * (2*(2+1))+3)*np.log(D)
print(FG)
```

```python
import statsmodels.api as sm
X = df[ ['rAdidas','rBayer']].dropna()
X = sm.add_constant(X)
y = df["DAX"].dropna()
modele2 = sm.OLS(y,X)
resultats2 = modele2.fit()
print(resultats2.summary())
```

```python
residus2=resultats2.resid
plt.plot(residus2)
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
```

```python
bp = het_breuschpagan(residus2, X)
print(bp)
```

```python
from statsmodels.stats.diagnostic import acorr_breusch_godfrey
```

```python
bg = acorr_breusch_godfrey(resultats2) 
print(bg)
```

```python
resultat_robust2 = resultats2.get_robustcov_results(cov_type='HAC', maxlags=10)
print(resultat_robust2.summary())
```
