# Introduction : rappels généraux <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p1'></a>

- Modèle de régression multiple :
    
$$ Y = X \beta + \epsilon $$ 
  
<br>
    
- Matrices  :
    - Le vecteur colonne de la variable expliquée : $ Y \in \mathcal{M}_{(T,1)}(\mathbb{R})$
    - La matrice des variables explicatives, dont la première colonne est composée de $1$ pour la constante : $ X \in \mathcal{M}_{(T,k+1)}(\mathbb{R})$
    - Le vecteur colonne des paramètres à estimer : $ \beta \in \mathcal{M}_{(k+1,1)}(\mathbb{R})$
    - Le vecteur colonne des termes d'erreur: $ \epsilon \in \mathcal{M}_{(T,1)}(\mathbb{R})$
    
<br>
    
- Hypothèses :
    - $X$ est non aléatoire : $E(\epsilon_t \vert X) = 0$ 
    - $X$ est de plein rang : $Rang(X) = k+1$
    - L'erreur est nulle en moyenne : $E(\epsilon) = 0$
    - Non autocorrélation et homoscédasticité des erreurs : $E(\epsilon \epsilon') = \sigma_{\epsilon}^2 I$
    - Normalité des erreurs : $\epsilon \sim N(0,\sigma_{\epsilon}^2I)$
    
<br>
    
- Résiuds : $e = Y - X\hat{\beta}$
    
<br>
    
- Estimateurs MCO : 
    - $\hat{\beta} = (X'X)^{-1}X'Y$
    - $\hat{\sigma}_{\epsilon}^2 = \frac{e'e}{T-k-1}$
   
<br>
    
- Propriétés : les estimateurs sont BLUE

<br>
    
- Tests d'hétéroscédasticité :
    - Breusch-Pagan : $BP \sim \chi_p^2$
        - $BP<\chi_p^2$, l'hypothèse nulle d'homoscédasticité n'est pas rejetée
        - $BP>\chi_p^2$, l'hypothèse nulle d'homoscédasticité est rejetée
    - Test de White : $TR^2 \sim \chi_{2k}^2$
        - si $TR^2 < \chi_{2k}^2$, on ne rejette pas l'hypothèse nulle d'homoscédasticité
        - si $TR^2 > \chi_{2k}^2$, on rejette l'hypothèse nulle d'homoscédasticité
    

- Tests d'autocorrélation :
    - Durbin-Watson : $ DW \simeq 2(1-\hat{\rho})$
        - On regarde le nombre d'observations et le nombre de variables explicative (hors constante)
        - On se sert du tableau de décision pour savoir si le test rejette ou non l'hypothèse nulle d'absence d'autocorrélation
    - Breusch-Godfrey : $BG \sim \chi^2_p$
        - $BG<\chi_p^2$, l'hypothèse nulle d'absence d'autocorrélation n'est pas rejetée
        - $BG>\chi_p^2$, l'hypothèse nulle d'absence d'autocorrélation est rejetée

    
<br>
    
Nous allons maintenant nous intéresser à l'hypothèse sur le fait que $X$ est non aléatoire et l'hypothèse de plein rang ($Rang(X) = k+1$), concernant les variables explicatives. Nous verrons aussi comment gérer l'éventualité de changements de structure dans les variables explicatives.

# Problème lié aux variables explicatives : variables aléatoires <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p1'></a>

- Hypothèse, $X$ est non aléatoire : $E(\epsilon_t \vert X) = 0$
    
    $\Rightarrow$ renvoie à l'hypothèse d'exogénéité des variables explicatives : l'espérance conditionnelle de l'erreur étant nulle, les variables explicatives ne sont pas déterminées par le modèle et lui sont donc exogène. 
    
    
- Pourquoi ? Permet la convergence de l'estimateur : $Plim (\hat{\beta}) = \beta$ lorsque $T\to +\infty$
    
<br>
    
- Lorsque cette hypothèse est non vérifée : même si la taille de l'échantillon croît indéfiniment, les estimateurs MCO ne se rapprochent plus de la vraie valeur.
- Problème : trouver un estimateur convergent
    
<br>
    
    
- L'idée : 
    - les variables explicatives ne sont pas réellement "exogènes"
    - solutions : les remplacer par des variables qui le sont
    - enjeux : trouver des variables exogènes, représentative de ce que l'on souhaite mesurer
  
<br>
    
- On utilise dans ce cas, la méthode des Variables Instrumentales
- Soit $Z$, une matrice de $k$ variables de sorte que :
    
$$ cov(Z'\epsilon) = 0$$
    
$$ cov(Z'X) \neq 0$$
    
<br>
    
    
- Alors il est possible d'estimer le modèle suivant par les MCO :
    
$$Z'Y = Z'X\beta + Z'\epsilon$$
   
<br>
    
- L'estimateur de ce modèle est appelé estimateur des variables instrumentales $\hat{\beta}_{VI}$
- $\hat{\beta}_{VI}$ vérifie les propriétés des estimateurs MCO et est BLUE.
    
    
<br>
    
<b> Détecter le problème</b>
    
- Pour détecter le problème : 
    - étude de la corrélation entre les variables explicatives et le résidu de la régression
    - mener un test d'Hausman (consiste à comparer $\hat{\beta}_{VI}$ et $\hat{\beta}_{MCO}$)
    
<br>
    
- En pratique, il est rare de recourir à cette méthode : difficile de trouver $Z$

    
<br>
    
<b> Origines du problème</b>
    
- Mauvaise mesure des variables
- Manque d'une variable explicative importante dans le modèle (fortement corrélée avec l'une des variables explicatives du modèle)
- La variable expliquée est aussi variable explicative du modèle
    
<br>
    
- Ce dernier problème renvoie à ce que l'on appelle généralement un problème <b>d'endogénéité</b> que nous étudierons plus en détail à la séance 10.
    
</span>

# Problème lié aux variables explicatives : multicolinéarité <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p2'></a>


<br>
    
    
Nous avons étudié l'hypothèse selon laquelle $X$ est une matrice non aléatoire. Nous nous intéressons maintenant, plus en détails à l'hypothèse qui établie que $X$ est de plein rang. 
    
    
</span>

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>
<hr style=" border: .5px solid black;">
   
## Présentation et effets de la multicolinéarité <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p2.1'></a>


- Rappel on fait l'hypothèse que $Rang(X) = k+1$, nombre de variables explicatives en plus de la constante.
- $\Rightarrow$ On ne peut écrire l'une des variables / colonnes comme une combinaison linéaire des autres : $\forall a_j \in \mathbb{R}$, et pour toute variable $i \in [\![1,k]\!]$
    
$$ X_i \neq \sum_{i\neq j}a_j \times X_j$$
    
    
- Cette hypothèse assure que $(X'X)$ est bien inversible
- Sans voir de multicolinéarité parfaite, il suffit que deux des variables  explicatives (ou plus) soient partiellement multicolinéaires pour que cela puisse altérer l'estimation de $\hat{\beta}$
    
    
<br>
    
- Formellement : la multicoinéarité parfaite signifie qu'il existe une famille $(\lambda_i)_{i \in [\![1,k]\!]}$ de constantes telle que :
    
$$\lambda_1 X_1 + ... + \lambda_k X_k = 0$$


- multicolinéarité imparfaite : 
    
$$\lambda_1 X_1 + ... + \lambda_k X_k +u = 0$$
    
où $u$ est un terme d'erreur. Dans ce cas, il sera potentiellement possible d'estimer $\hat{\beta}$ ($X'X$ inversible) mais avec des écarts-types très élevés (et donc des paramètres peu significatifs).
    
<br>
    
<b>Impact sur la précision des paramètres estimés :</b>

- Pour le modèle de régression linéaire multiple, on a : 
$$ Var(\hat{\beta}) = \sigma^2_{\epsilon}(X'X)^{-1}$$

- On peut montrer que : 
$$ Var(\hat{\beta}_j) = \frac{\sigma^2_u}{nVar(X_j)}\frac{1}{(1-R^2_j)}, \quad j \in {1, ..., k}$$

où $X_j$ désigne la $j^{ième}$ variable et $R^2_j$ le coefficient de détermination de la régression de $X_j$ sur une constante et les autres variables explicatives du modèle.

- En conséquence, si $X_j$ est liée aux autres variables par une relation linéaire non exacte mais forte, $R^2_j \rightarrow 1$ et $Var(\hat{\beta}_j) \rightarrow + \infty$

- Le paramètre $\beta_j$ est estimé avec peu de précision, ce qui donne peu de crédit à tout exercice d'inférence statistique l'impliquant.   
    
    
</span>

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>
<hr style=" border: .5px solid black;">
  

## Détection de la multicolinéarité <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p2.2'></a>

<br>
    
    
<b> Variance inflation factor</b>    
    
    
<br> 
    
Diagnostic: il vient naturellement qu’une statistique d’intérêt pour évaluer la présence de la multicolinéarité presque parfaite est égale à:

$$VIF_j = \frac{1}{1-R^2_j}, \quad 1,...,k$$

- <p style="text-align: justify;">Cette statistique est appelée facteur d’influence de la variance (Variance Inflation Factor).</p>

- <p style="text-align: justify;">Des valeurs élevées pour cette statistique pour une variable explicative donnée est le signe de la présence éventuelle d’une situation de multicolinéarité presque parfaite l’impliquant.</p>

- <p style="text-align: justify;">Il n’existe cependant pas une règle établissant un seuil pour cette statistique au-delà duquel on peut prendre une décision. La statistique est néanmoins utile car donnant des indications sur les paramètres $\beta_j$ pouvant être estimés avec peu de précision.</p>
    
- Parfois considéré que : $VIF<2$ est parfait, et qu'à partir de $10$, la multicolinéarité est trop forte.
    
<br>
    
    
<b> Corrélation entre variables eplicatives </b>
    
    
<br>     
    

    
    
    
    
<br>
    
    
<b> Test de Farrar Glauber</b>    
    
    
<br> 
    
- Test basé sur la matrice des coefficients de corrélation entre les variables explicatives.
- Si les variables sont linéairement corrélées, le déterminant de cette matrice est nul
- La statistique de test est données par :
    
$$FG = -(T-1-\frac{1}{6}(2(k+1)+5))\times log(D)$$
    
où $D$ est le déterminant de la matrice de corrélation des variables explicatives.
    
- Sous l'hypothèse nulle d'orthogonalité (non-multicolinéarité) des variables explicatives : 
    
$$FG \sim \chi_{1/2k(k+1)}^2$$

où la table du $\chi^2$ est donnée ici : <a href="https://people.smp.uq.edu.au/YoniNazarathy/stat_models_B_course_spring_07/distributions/chisqtab.pdf">Tables du chi2</a>
    
- La règle de décision est données par : 
    - Si $FG <\chi_{\frac{1}{2}k(k+1)}^2$, l'hypothèse nulle n'est pas rejetée (pas de multicolinéarité)
    - Si $FG >\chi_{\frac{1}{2}k(k+1)}^2$, l'hypothèse nulle est rejetée (présomption de multicolinéarité)    
    
    

    
</span>

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>
<hr style=" border: .5px solid black;">
  

## Solutions à la multicolinéarité <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p2.3'></a>


- <b>Augmenter le taille de l'échantillon</b> : augmenter le nombre d'observations pour diminuer la part des observations corrélées entre elles. Plus $T$ est supérieur à $k$, moins la multicolinéarité a de chance d'être forte.
    - $\Rightarrow$ problème : si les variables sont "par définition" colinéaires

<br>
    
- <b>La régression ridge</b> : 
    - l'une des colonnes de $X$ est égale à une combinaison linéaire des autres 
    - en ajoutant une constante sur la diagonale de $X$, cette multicolinéarité disparaît

$$\hat{\beta}_R = (X'X + cI)^{-1}X'Y$$
    
<br>
    
- Difficulté : choix de $c$ (résolution algorithmique de l'hyperparamètre).
    
<br>
    
- Note : l'estimateur par la régression ridge n'est plus sans biais, mais il permet d'obtenir un variance moindre (donc estimation plus précise).


- <b>Autres méthodes</b> :
    - suppréssion des varibles explicatives à l'origine de la multicolinéarité
    - transformation des variables (en taux de croissance par exemple, ou en logarithme)
    - analyse en composante principale.
    - ...



</span>

## Méthodes de sélection de variables <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p2.4'></a>

Ces méthodes servent à déterminer les variables explicatives à garder dans un modèle lorsque l'on est en présence de multicolinéarité et que l'on souhaite en retirer. 
    
    
- Trois manières de procéder principalement : 
    - on teste toutes les régressions possibles (et on compare les résultats avec les critères de comparaison de modèles)
    - <i>backward elimination</i> : on estime le modèle complet puis on retire les variables ayant la t-stat la plus faible. On retient ensuite le meilleur modèle
    - <i>forward regression</i> : on estime le modèle avec une seule variable (celle dont la corrélation est la plus forte avec la variable expliquée), puis on en ajoute une autre, etc... On retient le meilleur modèle.

## L'introduction de variables indicatrices <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3.1'></a>

- Variables indicatrices (dummy) : variables binaire composées de $0$ et de $1$.
- Utilisées pour caractériser la présence/absence d'un phénomène (individus ou évènement).
- On peut donc introduire les dummy comme variables explicatives dans un modèle :
    - elle vaut $0$ pour les années où le phénomène n'a pas lieu
    - elle vaut $1$ pour l'année (les années) où le phénomène à lieu
- On parle aussi de variable qualitative.
    
<br>
    
- Exemple : 
    
$$Y_{t} = \alpha + \delta_1 D_t + \beta_1 X_{1,t} + ... + \beta_k X_{k,t} + \epsilon_t$$
  
<br>
    
- Quelques exemples :
    - taux de criminalité : plus élevé en zone urbaine ou rurale ?

- Interraction entre variables : dummy et variables explicatives

$$Y_{t} = \alpha + \delta_1 D_t + \beta_1 X_{1,t} + \beta_2 ( X_{1,t} \times D_t)  + \epsilon_t$$
    
    

<br>
    
    
- Interprétation : 
    - $\delta_1$ renvoie à l'effet d'être concerné par cette dummy (GSIB ou DSIB dans notre exemple)
    - $\beta_1$ l'effet de la variable $1$ sur la variable expliquée (le ratio de levier par exemple)
    - $\beta_2$ est l'effet de la variable $X_1$ lorsque l'individu ou la période est conernée par $D$ (effet du ratio de levier pour une banque systémique dans notre exemple)
    
<br>
    
    
</span>

## Tests de stabilité des coefficients <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3.2'></a>


- En robustesse de l'estimation menée : tester la stabilité des coefficients (en série temporelle principalement)
    - S'assurer que les coefficients ne varient pas selon le contexte
    - Etudier l'impact d'un changement structurel
    
<br>
    
- Parmis d'autres approches, nous étudions ici le test de Chow
    
<br>
    
- Prenons le modèle de régression multiple :
    
    
$$Y_t = \beta' X_t + \epsilon_t$$
    
pour tous les $t = 1,...,T$.
    
<br>    
    
    
- On divise en deux l'échantillon (en deux périodes par exemple) :
    
$$Y_{t} = \alpha_1 + \beta_1 X_t + \epsilon_{1,t}$$
    
pour les $t = 1,...,\tau$
    
$$Y_{t} = \alpha_2 + \beta_2 X_t + \epsilon_{2,t}$$
    
    
pour les $t = \tau +1 ,...,T$
    
  
- On choisi $\tau$ en fonction du changement structurel que l'on souhaite étudier (une crise ou un changement de législation ou de monnaie par exemple)   
    
<br>
    
- L'idée est de tester : $H_0$ : $\alpha_1 = \alpha_2$ et $\beta_1 = \beta_2$
    

- Pour cela, on estime les trois modèles et on récupère la somme des carrés des résidus des trois modèles ($SCR_1$, $SCR_2$ et $SCR_3$).
- On pose $SCR_a = SCR_2 + SCR_3$
- Alors :
    
   
$$F = \frac{(SCR_1 - SCR_a)/(k+1)}{SCR_a/(T-2(k+1)} \sim F(k+1,T-2(k+1)) $$
    
<br>
    
- Règle de décision :
    - Si $F < F(k+1,T-2(k+1))$, non rejet de $H_0$ : il n'y a pas de changement structurel
    - Si $F > F(k+1,T-2(k+1))$, rejet de $H_0$ : il y a changement structurel
    
    
<br>
    
Exemple d'application en python : https://www.statology.org/chow-test-in-python/
    
    
<br>
