Afin de contr√¥ler l'effet des √©v√©nements ayant impact√© l'√©conomie allemande et europ√©enne au cours des ann√©es pr√©c√©dentes, nous introduirons deux variables dummy pour indiquer s'il s'agit d'une p√©riode de crise sanitaire ou de guerre.

```python
import warnings

# Ignorer les avertissements
warnings.filterwarnings('ignore')


# Assurer que la colonne 'Date' est au format datetime
df['Date'] = pd.to_datetime(df['Date'])

# D√©finir les p√©riodes pour chaque √©v√©nement
# P√©riode de COVID-19 : d√©but en mars 2020
covid_start = '2020-03-01'
covid_end = '2020-09-01'  # Fin approximative des mesures strictes


# P√©riode de la guerre russo-ukrainienne : d√©but en f√©vrier 2022
war_start = '2022-02-01'
war_end = '2022-12-31'  # √Ä ajuster selon les d√©veloppements

# Cr√©er les colonnes dummy
df['COVID'] = ((df['Date'] >= covid_start) & (df['Date'] <= covid_end)).astype(int)
df['War'] = ((df['Date'] >= war_start) & (df['Date'] <= war_end)).astype(int)
```

#Commentaire1:

#Commentaire2:

la correlation entre rDAX et War n√©gative mais faible (-0.028456 soit -2,84%) ainsi que entre rAdidas et War (-0.052421 soit -5,24%) La pr√©sence de guerre semble avoir une influence n√©gative mais faible sur ces deux variable

la correlation entre rDAX et COVID tr√®s faible (0.014342 soit 1,43%) ainsi que entre rAdidas et COVID (0.005061 soit 0.005%) La pr√©sence de guerre semble avoir une influence n√©gative sur ces deux variable

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>


    
<br>

Nous estimons le mod√®le de base. <br>    
$$
                              rAdidas= \alpha + \beta_1 \cdot rDAX + \beta_2 \cdot War + \beta_3  \cdot Covid + \epsilon  
$$
- Commentez de fa√ßon pr√©cise les r√©sultats (significativit√©, qualit√© globale de l'ajustement).<br>
- Commentez le signe et l'ampleur des coefficients significatifs.

<br>
</span>

```python
import statsmodels.api as sm

df = df.dropna()
X = df[["rDAX", "War", "COVID"]]
X = sm.add_constant(X)

y = df["rAdidas"] # A MODIFIER !!!!!!!!!!!!!

modele = sm.OLS(y,X)

resultats = modele.fit()

print(resultats.summary())
```

Le $R^2$ est de 0.432, c'est plut√¥t une bonne qualit√© d'ajustement. Cela signifie que 43,2% de la variance de la rentabilit√© d'Adidas ($rAdidas$) est expliqu√© par ce mod√®le 

Nous pouvons faire un test de significativit√© globale de Fischer pour savoir si le mod√®le est significatif dans son  ensemble. On compare alors la p-value au seuil critique de 5% 
$$
\text{Prob (F-statistic): } 6.31e-155 <0.05
$$
Le mod√®le est donc globalement significatif au sens du Fischer

Dans ce mod√®le un seul des coefficient est significatif au sens du Student si on regarde les p-value et qu'on les compare au seuil critique de 5%. 
$$
\text{p-value}\;rDAX =   0.000
$$
On estime que lorque le DAX progresse de 1% Adidas progresse de 1.1276%, toutes choses √©tant √©gales par ailleurs.

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>


    
<br>
    
<b>R√©sidus :</b>  <br>
 Faites un graphique ligne des r√©sidus. **Commentez.**

<br>
</span>

```python
#R√©sidus du mod√®le
residus = resultats.resid

#Graphique
residus.plot()
```

La variance n'apparait pas constante c'est a dire que l'on peut regrouper des individus par groupe avec ceux qui ont des petites variance et ceux qui ont une grosse variance. Cette anomalie prouve nous avons donc une pr√©somption d'h√©t√©rosc√©dasticit√©

La pr√©sence de cluster dans les r√©sidus peuvent √©galement illustrer de l'autocorrelation

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>


    
<br>
    
Afin de v√©rifier les hypoth√®ses classiques de l'estimation en MCO, nous effectuons les tests d'homosc√©dasticit√© et d'autocorr√©lation.
 
 **Commentez les r√©sultats des tests.**

<br>
</span>

```python
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
from statsmodels.stats.diagnostic import acorr_breusch_godfrey
```

```python
# Test h√©t√©roscedasticit√© BP
stat1 = het_breuschpagan(residus, X)
print(stat1[1])
```

La p-value (0.0026801156215007925) est inf√©rieure ou √©gale √† 0,05 : On rejette $H0$, selon le test de Breush Pagan nous sommes en pr√©sence d'h√©t√©rosc√©dasticit√© 

```python
# Test h√©t√©roscedasticit√© W
stat2 = het_white(residus, X)
print(stat2[1])
```

La p-value (0.00010214730340947128) est inf√©rieure ou √©gale √† 0,05 : On rejette $H0$, le test de White vient confirmer nos doutes, nous sommes en pr√©sence d'h√©t√©rosc√©dasticit√© 

```python
#Test d'autocorr√©lation
stat3 = acorr_breusch_godfrey(resultats, nlags=10)
print(stat3[1])
```

La p-value 0.003661828841230082 <  0.05 on rejette H0, nous sommes en pr√©sence d'autocorr√©lation

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>


    
<br>
    
<b>Correction: </b>  <br>
Pour rem√©dier √† ces probl√®mes, la m√©thode d'estimation de Newey-West apporte une correction aux estimateurs. Nous l'appliquons ci-dessous.

 **Commentez les r√©sultats**

<br>
</span>

```python
resultat_robust = resultats.get_robustcov_results(cov_type='HAC', maxlags=10)
print(resultat_robust.summary())
```

On retrouve les m√™mes estimations avec les p-value inchang√© ce qui n'est pas normal. En th√©orie apr√®s la correction de newey west les autres variable devrait devenir significatives, ce n'est pas le cas. Le probl√®me est donc plus profond. Il est sur√™ment li√© √† la sp√©cification du mod√®le : soit nous avons mal choisit nos variables, soit l'utilisation d'un mod√®le lin√©aire n'est pas adapt√© (le coefficient de corr√©lation entre radidas & rdax est √©l√©v√© ce serait donc √©tonnant) √† l'√©tude √©conom√©trique que nous voulons faire.
