```python
import yfinance as yf
import pandas as pd 
import matplotlib.pyplot as plt 
import warnings
import seaborn as sns

#ignore all warning 

warnings.filterwarnings('ignore')
```

```python
tickers = ['TSLA', 'BTC-USD', 'CL=F', 'DX-Y.NYB','^GSPC',"^IXIC",'NVDA', '^TNX','NG=F']
start_date = '2018-01-01'
end_date = '2024-11-06'
data = yf.download(tickers, start = start_date, end = end_date, group_by = tickers)
data.head()
```

```python
df = data.xs('Adj Close', level=1, axis= 1)

df.head()
```

```python
return = 100*df.pct_change().dropna()
```
