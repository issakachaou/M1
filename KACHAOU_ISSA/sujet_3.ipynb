# Echauffement
    
<br>
    
1- On nous donne un dictionnaire avec deux clefs (`var1` et `var2`) ayant chacune pour valeur une liste d'integer ou float. Créez un troisième élément avec comme clef `somme` et prennant comme valeur, la somme des deux premières clefs (pas de fonction).
    
<br>

````python
# Initialisation du dictionnnaire

d = {"var1":[1,2,3,4], "var2":[1.1,2.5,8.3,-12]}

# Création d'une nouvelle clé (keys) qui prend le nom "var3" et qui somme les valeurs associés au clé "var1" et "var2"

d["var3"]= d["var1"]+d["var2"]

# Affichage du nouveau dictionnaire

print(d)
````

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>
    
<br>
    
2- A l'aide de votre réponse à la question précédente, proposez une fonction permettant de calculer cette troisème variable `somme` pour n'importe quel dictionnaire contenant déjà deux variables.
    
<br>

````python
def somme(X,Y):
    """
    Cette fonction fait la somme de deux liste de valeur contenue dans un dictionnaire
    Variable : 
    X : list
    Y : list
    Dépendance : aucune 
    """
    # Fait la somme et stock la valeur dans la variable Z
    Z = X + Y
    # Retourne Z
    return Z
````

````python
# Test de la fonction avec une nouvel exemple de dictionnaire
d = {"A":[1,2,3,4], "B":[1.1,2.5]}

somme(d["A"],d["B"])
````

# Exercice thématique - Echec et mat
    
<br>
    
<b>Objectif </b>: on nous donne un dictionnaire avec en clefs des pièces d'échec et en valeur leur position. On nous donne également une liste de mouvements de pièces. L'objectif est de modifier le dictionnaire de sorte qu'il prenne en compte les mouvements effectués. Les mouvements sont toujours construits de la même manière : chaine de caractères avec en première lettre la pièce qui bouge, puis la case vers laquelle la pièce se dirige. 
    
<br>

    
Pour information : 
    
- il n'y a pas besoin de savoir jouer aux échecs pour comprendre et résoudre l'énoncé. 
    
- un jeu d'échec est composé de 8x8 cases. Chaque colonne est identifiée par une lettre (de `a` à `h`) et les lignes par des numéro (de `1` à `8`). Ainsi la case tout en bas à gauche est la case `a1`.  
    
- les pièces sont identifiées comme tel : majuscule pour les blancs / minuscules pour les noirs. `K` pour le roin `Q` pour la dame, `N` pour le cavalier, `B` pour le fou, `R` pour la tour, `P` pour le pion.
    
    
<b>Exemple 1</b> :
    
```python
d_chess = {"K":"e1", "k":"e8", "Q":"d1"}
moves = ["Qa4"]
```
deviendra :
```python
{"K": "e1", "k": "e8", "Q": "a4"} 
```
 

    
<b>Exemple 2</b> :
    
```python
d_chess = {"K":"e1", "k":"e8", "Q":"d1", "p1":"a5"}
moves = ["Qa4", "ke7","Qe4"]
```
deviendra :
```python
{"K": "e1", "k": "e7", "Q": "e4", "p1":"a5"} 
```
<br>

<br>
    
    
<b>Exemple 3</b> :
    
