# Introduction : rappels sur le modèle de régression simple <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p1'></a>

- Modèle de régression <b>linéaire</b> dans les paramètres


- Modèle simple : quand il n'y a qu'une seule variable explicative


- Hypothèses sur le terme d'erreur : 
     - nulité en moyenne ($  E (\epsilon_t ) = 0 $)
     - absence d'autocorrélation du terme d'erreur ($ E (\epsilon_t \epsilon_{t'}) = 0 , \forall t \neq t'$ )
     - homoscedasticité du terme d'erreur ($  E (\epsilon_t^2) = \sigma^2_{\epsilon} , \forall t $)
     - normalité du terme d'erreur ($\epsilon_t \sim N(0,\sigma^2_{\epsilon})$)
     
     
- Principe des MCO : minimiser la sommes des carrés des résidus ($ MCO \Longleftrightarrow Min \sum_{t=1}^T e_t^2$)


- Résultats MCO : :
    - $\hat{\alpha} = \bar{Y} - \hat{\beta} \bar{X} $
    - $ \hat{\beta} = \frac{Cov(X_t,Y_t)}{V(X_t)}$
    
    
- Propriété des estimateurs MCO (parmi d'autres) : ils sont BLUE (Best Linear Unbiased Estimator).

# Ecriture sous forme matricielle <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p2'></a>


Nous passons au modèle de régression multiple, c'est à dire au modèle dans lequel il y a plusieurs variables explicatives. Ainsi, le modèle de régression multiple à $k$ variables s'écrit :

$$ Y_t = \alpha + \beta_1 X_{1,t} + \beta_2 X_{2,t} + ... + \beta_k X_{k,t} + \epsilon_t $$

où $t \in [\![1,T]\!]$, $Y_t$ est la variable expliquée, $\{X_{i,t}, i \in [\![1,k]\!]\}$ sont les $k$ variables explicatives et $\epsilon_t$ est le terme d'erreur.

Les coefficients $\beta_i$ sont les paramètres que l'on cherche à estimer. Ils correspondent à <b>la variation de la valeur moyenne de $Y$</b>, lorsque les autres variables sont constantes ("toutes choses égales par ailleurs"). 

On peut réécrire le modèle de la façon suivante, pour chaque valeur de $t$ :

$$ Y_1 = \alpha + \beta_1 X_{1,1} + \beta_2 X_{2,1} + ... + \beta_k X_{k,1} + \epsilon_1 $$
$$ Y_2 = \alpha + \beta_2 X_{1,2} + \beta_2 X_{2,2} + ... + \beta_k X_{k,2} + \epsilon_2 $$
$$ ... $$
$$ Y_T = \alpha + \beta_1 X_{1,T} + \beta_2 X_{2,T} + ... + \beta_k X_{k,T} + \epsilon_T $$

ou encore sous la forme matricielle :

$$ Y = X \beta + \epsilon $$ 

où :

$$Y = 
\begin{pmatrix}
Y_1 \\
Y_2 \\
\vdots  \\
Y_T \\
\end{pmatrix}
$$

$$X = 
\begin{pmatrix}
1 & X_{1,1} & X_{2,1} & ... & X_{k,1}\\
1 & X_{1,2} & X_{2,2} & ... & X_{k,2}\\
\vdots & \vdots & \vdots & ... & \vdots\\
1 & X_{1,T} & X_{2,T} & ... & X_{k,T}\\
\end{pmatrix}
$$


$$\beta = 
\begin{pmatrix}
\alpha \\
\beta_1 \\
\beta_2 \\
\vdots  \\
\beta_k \\
\end{pmatrix}
$$

et

$$\epsilon = 
\begin{pmatrix}
\epsilon_1 \\
\epsilon_2 \\
\vdots  \\
\epsilon_T \\
\end{pmatrix}
$$

donc on a des matrices de dimensions :

- Le vecteur colonne de la variable expliquée : $ Y \in \mathcal{M}_{(T,1)}(\mathbb{R})$
- La matrice des variables explicatives, dont la première colonne est composée de $1$ pour la constante : $ X \in \mathcal{M}_{(T,k+1)}(\mathbb{R})$
- Le vecteur colonne des paramètres à estimer : $ \beta \in \mathcal{M}_{(k+1,1)}(\mathbb{R})$
- Le vecteur colonne des termes d'erreur: $ \epsilon \in \mathcal{M}_{(T,1)}(\mathbb{R})$

# Estimateur MCO <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3'></a>


De même que pour le modèle de régression simple, nous avons un certain nombre d'hypothèses qu'il nous faut respecter pour pouvoir mener une estimation MCO dans le cadre d'une régression multiple. 



## Hypothèses du modèle de régression multiple <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3.1'></a>

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<b>Hypothèse H1 : </b> la matrice $X$ est non aléatoire

Les valeurs de $X$ sont observées sans erreur. Une manière moins contraignante d'exprimer cette hypothèse est de dire que $X$ et le terme d'erreur sont indépendants : 

$$E(\epsilon_t \vert X) = 0$$

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>Hypothèse H2 : </b> la matrice $X$ est de plein rang

Mathématiquement : $Rang(X) = k+1$ (le rang de la matrice est égale à son nombre de colonnes, c'est à dire au nombre de variable plus la constante). La dimension du sous-espace vectorielle engendré par les vecteurs colonnes  de la matrice est égal an nombre de colonnes. 

Cette hypothèse revient à dire que les colonnes sont linéairement indépendantes, ou que l'on ne peut pas écrire une colonne de $X$ comme une combinaison linéaire des autres. 

Nous faisons cette hypothèse pour assurer l'inversibilité de la matrice $X'X$ (dont nous avons besoin pour le calcul de $\beta$).

Une hypothèse sous-jacente est que le nombre d'observations ne peut pas être inférieur au nombre de variables : $T > k+1$. Sans cela, $X$ ne peux pas être de plein rang. 




<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>


<b>Hypothèse H3 : </b> nulité de l'espérance de l'erreur

Comme pour le modèle de régression simple, nous supposons que le terme d'erreur est nul en moyenne :

$$ E(\epsilon) = 0$$

On en déduit que :

$$ E(Y) = X \beta $$

c'est à dire que le modèle est en moyenne bien spécifié.

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>Hypothèse H4 : </b> homoscédasticité et absence d'autocorrélation des erreurs


Soit $E(\epsilon \epsilon')$ la matrice de variance covariance du terme d'erreur, avec $\epsilon'$ la transposée du vecteur des termes d'erreur. 

L'hypothèse H4 s'écrit comme suit : 

$$E(\epsilon \epsilon') = \sigma_{\epsilon}^2 I$$ 

où $I$ désigne la matrice identitée et $\sigma_{\epsilon}^2$ la variance du terme d'erreur.

En reprenant l'écriture matricielle de la matrice de variance-covariance, nous avons :

$$E(\epsilon \epsilon') = 
\begin{pmatrix}
V(\epsilon_1) & Cov(\epsilon_1,\epsilon_2) & Cov(\epsilon_1,\epsilon_3) & ... & Cov(\epsilon_1,\epsilon_T)\\
Cov(\epsilon_2,\epsilon_1) & V(\epsilon_2)  & Cov(\epsilon_2,\epsilon_3) & ... & Cov(\epsilon_2,\epsilon_T)\\
\vdots & \vdots & \vdots & ... & \vdots\\
Cov(\epsilon_T,\epsilon_1) & Cov(\epsilon_T,\epsilon_2) & Cov(\epsilon_T,\epsilon_3) & ... & V(\epsilon_T) \\
\end{pmatrix}
$$

$$ = 
\begin{pmatrix}
E(\epsilon_1^2) & E(\epsilon_1\epsilon_2) & E(\epsilon_1\epsilon_3) & ... & E(\epsilon_1\epsilon_T)\\
E(\epsilon_2\epsilon_1) & E(\epsilon_2^2)  & E(\epsilon_2\epsilon_3) & ... & E(\epsilon_2\epsilon_T)\\
\vdots & \vdots & \vdots & ... & \vdots\\
E(\epsilon_T\epsilon_1) & E(\epsilon_T\epsilon_2) & E(\epsilon_T\epsilon_3) & ... & E(\epsilon_T^2) \\
\end{pmatrix}
$$

L'hypothèse d'homoscédasticité et d'absence d'autocorrélation revient donc à dire que tous les termes de covariance sont nuls et que les termes de variance sont constants entre les observations :

$$E(\epsilon \epsilon') = 
\begin{pmatrix}
\sigma_{\epsilon}^2  & 0 & 0 & ... & 0\\
0 & \sigma_{\epsilon}^2   & 0 & ... & 0\\
\vdots & \vdots & \vdots & ... & \vdots\\
0 & 0 & 0 & ... & \sigma_{\epsilon}^2  \\
\end{pmatrix}
$$

$$ = \sigma_{\epsilon}^2 
\begin{pmatrix}
1  & 0 & 0 & ... & 0\\
0 & 1   & 0 & ... & 0\\
\vdots & \vdots & \vdots & ... & \vdots\\
0 & 0 & 0 & ... & 1  \\
\end{pmatrix}
$$


Comme pour le modèle de régression simple, nous appelons cette hypothèse, <b>l'hypothèse de sphéricité</b> des erreurs. 

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>


<b>Hypothèse H5 : </b> normalité des erreurs

Là encore, comme pour le modèle de régression simple, nous supposons (en vertu du théorème central limite) que les erreurs suivent une loi normale d'espérance nulle et de variance constante :

$$\epsilon \sim N(0,\sigma_{\epsilon}^2I)$$

À noter que cette hypothèse ne sert pas au calcul des paramètres mais à la validation des résultats statistiques et à la construction des tests statistiques.

## Estimation des cofficients <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3.2'></a>


De la même manière que pour le modèle de régression simple, nous cherchons à résoudre le programme suivant (minimisation de la somme des carrés des résidus) en notant $e = Y - X \hat{\beta}$ le vecteur des résidus : 

$$Min_{\hat{\beta}} \sum_{t=1}^T e^2 = Min_{\hat{\beta}} (e'e) \\
 = Min_{\hat{\beta}} (Y - X \hat{\beta})'(Y - X \hat{\beta}) $$

En effet : 

$$(e'e) = 
\begin{pmatrix}
e_1 & e_2 & ... &  e_T \\
\end{pmatrix}
\begin{pmatrix}
e_1 \\
e_2 \\
\vdots  \\
e_T \\
\end{pmatrix}
$$

$$ = e_1 \times e_1 + e_2 \times e_2 + ... + e_T \times e_T $$


Le résultat de cette minimisation est donné par : 


$$\boxed{  \hat{\beta} = (X'X)^{-1}X'Y  } $$


<br>

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">


<b>Preuve</b> :

En développant la somme des carrés des résidus, on obtient :

$$e'e = Y'Y -  2\hat{\beta}'X'Y +  \hat{\beta}'X'X \hat{\beta}$$

<i>Remarque</i> : en regardant de plus prêt les dimensions des matrices et vecteurs ci-dessus, on constate que chaque terme de la somme constitue bien un scalaire.

$$ \frac{\delta (e'e)}{\delta  \hat{\beta}} = 0 $$

avec 

$$ \frac{\delta (e'e)}{\delta  \hat{\beta}} = -2  \frac{\delta (\hat{\beta}X'Y)}{\delta \hat{\beta}} + \frac{\delta(\hat{\beta}'X'X\hat{\beta})}{\delta \hat{\beta}}$$

ce qui donne, en appliquant les propriétés de dérivation matricielle :

$$ \frac{\delta (e'e)}{\delta  \hat{\beta}} = -2X'Y + 2(X'X)\hat{\beta} = 0$$

$$(X'X)\hat{\beta} = X'Y $$

Et, par hypothèse <b>H2</b>, comme $X$ est de plein rang, $X'X$ est inversible. D'où :

$$  \hat{\beta} = (X'X)^{-1}X'Y $$

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

En python, qu'est ce que ça pourrait donner ? 

</span>

```python
# nous allons procéder à l'estimation de beta en utilisant la formule
# pour cela nous avons besoin de numpy
import numpy as np
```

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>

On procède au calcul de : 

$$X'X$$



</span>

```python
# la fonction np.transpose de numpy permet d'avoir la transposé d'une matrice

# on utilise un @ pour multiplier deux matices (il y a aussi la possibilité d'utiliser la fonction .dot())

XX = np.transpose(X) @ X
```

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>

Puis

$$(X'X)^{-1} $$



</span>

```python
# la fonction np.linalg.inv permet d'inverser la matrice
XX_1 = np.linalg.inv(XX)
```

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>

Il ne nous reste plus qu'à multiplier par $X'Y$ pour obtenir $\hat{\beta}$

$$  \hat{\beta} = (X'X)^{-1}X'Y $$



</span>

```python
# on réutilise la fonction np.transpose pour la transposée et .dot() pour le produit matriciel
params = XX_1.dot(np.transpose(X)).dot(y)
print(params)
```

## Propriétés des estimateurs MCO <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3.3'></a>


<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>Propriété 1</b> :  linéarité de l'estimateur

$$  \hat{\beta} = (X'X)^{-1}X'Y\\
= (X'X)^{-1}X'(X\beta + \epsilon)\\
= (X'X)^{-1}X'X\beta + (X'X)^{-1}X'\epsilon\\
= \beta + (X'X)^{-1}X'\epsilon$$


On a bien $\hat{\beta}$ qui est un estimateur linéaire de $\beta$.


<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>Propriété 2</b> : estimateur sans biais

Nous avons obtenu que :

$$  \hat{\beta} = \beta + (X'X)^{-1}X'\epsilon$$

Ainsi,

$$  E(\hat{\beta}) = E(\beta) + (X'X)^{-1}X'E(\epsilon)$$

Or, $E(\beta) = \beta$ et $E(\epsilon) = 0$, alors :

$$\boxed{E(\hat{\beta}) = \beta }$$

ce qui signifie que $\hat{\beta}$ est un estimateur sans biais de $\beta$.


<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>Propriété 3</b> : $\hat{\beta}$ est de variance minimale


Soit $\Omega_{\hat{\beta}}$ la matrice de variance-covariance de l'estimateur :




$$\Omega_{\hat{\beta}} = E[(\hat{\beta} - \beta)(\hat{\beta} - \beta)'] $$



$$\Omega_{\hat{\beta}} = 
\begin{pmatrix}
V(\hat{\alpha}) & Cov(\hat{\alpha},\beta_1) & Cov(\hat{\alpha},\beta_2) & ... & Cov(\hat{\alpha},\hat{\beta}_k)\\
Cov(\hat{\beta}_1,\hat{\alpha}) & V(\hat{\beta}_1)  & Cov(\hat{\beta}_1,\hat{\beta}_2) & ... & Cov(\hat{\beta}_1,\hat{\beta}_k)\\
Cov(\hat{\beta}_2,\hat{\alpha}) & Cov(\hat{\beta}_2,\hat{\beta}_1) & V(\hat{\beta}_2) & ... & Cov(\hat{\beta}_2,\hat{\beta}_k)\\
\vdots & \vdots & \vdots & ... & \vdots\\
Cov(\hat{\beta}_k,\hat{\alpha}) & Cov(V(\hat{\beta}_k,\hat{\beta}_1) & Cov(V(\hat{\beta}_k,\hat{\beta}_2) & ... & V(\hat{\beta}_k) \\
\end{pmatrix}
$$

<br>

Comme on sait que $  \hat{\beta} = \beta + (X'X)^{-1}X'\epsilon$, on peut réécrire la matrice de variance covariance des estimateurs :


$$\Omega_{\hat{\beta}} = E \left([(X'X)^{-1}X'\epsilon][(X'X)^{-1}X'\epsilon]'\right)$$


et, puisque la transposée d'une matrice symétrique est égale à la matrice elle-même :

$$\Omega_{\hat{\beta}} = E \left((X'X)^{-1}X'\epsilon\epsilon'X(X'X)^{-1}\right)$$


et comme $E(\epsilon\epsilon') = \sigma_{\epsilon}^2I$, nous obtenons,


$$\Omega_{\hat{\beta}} = (X'X)^{-1}X'\sigma_{\epsilon}^2I'X(X'X)^{-1}$$

et enfin,

$$ \boxed{\Omega_{\hat{\beta}} = \sigma_{\epsilon}^2(X'X)^{-1}}$$


Il est possible de montrer que tout autre estimateur de $\beta$ (autre que celui des MCO) sera de variance supérieure à cette variance. Autrement dit, l'etimateur MCO est de variance minimale.



<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>Propriété totale</b> : $\hat{\beta}$ est un estimateur BLUE de $\beta$


On dit de l'estimateur MCO $\hat{\beta}$ de $\beta$ est <b>BLUE</b> (Best Linear Unbiased Estimator).

## Estimateur de la variance des erreurs <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p3.4'></a>


On ne connait pas les erreurs dans un modèle. Pour estimer la variance des erreurs $\sigma_{\epsilon}^2$, il convient d'utiliser les résidus $e$ :

$$ \boxed{ \hat{\sigma}_{\epsilon}^2 = \frac{e'e}{T-k-1}}$$


$\hat{\sigma}_{\epsilon}^2$ est un estimateur sans biais de la variance des erreur $\sigma_{\epsilon}^2$.

## Loi des estimateurs <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p4.1'></a>


Nous utilisons à la fois l'hypothèse <b>H5</b>, qui suppose que les erreurs suivent une loi normale, et l'estimateur de la variance des erreurs pour obtenir que :

$$\frac{(T-k-1)\hat{\sigma}_{\epsilon}^2}{\sigma_{\epsilon}^2} \sim \chi^2_{T-k-1} $$

Où la loi du $\chi^2$ est obtenu par somme de variables aléatoires suivant une loi normale. De là nous pouvons déduire que :

$$t = \frac{\hat{\beta}_i - \beta_i}{\hat{\sigma}_{\epsilon} \sqrt{a_{i+1,i+1}}} \sim t(T-k-1)$$

où $i$ désigne la $i$ème variable de $X$ et $a_{i+1,i+1}$ le $i+1$ème élément de la diagonale de $(X'X)^{-1}$.

</span>

## Tests sur un coefficient de régression : test de Student <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p4.2'></a>

À partir de la statistique $t$ définie à la section précédente, nous pouvons déduire un intervalle de confiance à $100(1-p)\%$ pour $\beta_i$ :

$$\hat{\beta}_i \pm t_{p/2}\hat{\sigma}_{\epsilon} \sqrt{a_{i+1,i+1}}$$

On peut donc tester si $\hat{\beta}_i$ s'écarte plus ou moins une certaine valeur $\beta_0$ : 

- $H_0$, l'hypothèse nulle : $\hat{\beta}_i = \beta_0$
- $H_1$, l'hypothèse alternative : $\hat{\beta}_i \neq \beta_0$

si l'hypothèse nulle est vraie, alors :

$$\frac{\hat{\beta}_i - \beta_0}{\hat{\sigma}_{\epsilon} \sqrt{a_{i+1,i+1}}} \sim t(T-k-1)$$


Règle de décision :
- Si $\vert \frac{\hat{\beta}_i - \beta_i}{\hat{\sigma}_{\epsilon}^2 \sqrt{a_{i+1,i+1}}} \vert \leq t_{p/2}$, on ne rejette pas l'hypothèse nulle au seuil statistique de $100p\%$ et on considèe que $\hat{\beta}_i = \beta_0$
- Si $\vert \frac{\hat{\beta}_i - \beta_i}{\hat{\sigma}_{\epsilon}^2 \sqrt{a_{i+1,i+1}}} \vert > t_{p/2}$, on rejette l'hypothèse nulle au seuil statistique de $100p\%$ et on considèe que $\hat{\beta}_i \neq \beta_0$

<hr style=" border:none;border-top:.5px dotted black;color:black;background-color:#fff;height:.5px;">

<br>

<b>En pratique</b>, on utilise ce test comme un test de significativité, c'est à dire qu'on prend $\beta_0 = 0$ :

- $H_0$, l'hypothèse nulle : $\hat{\beta}_i = 0$
- $H_1$, l'hypothèse alternative : $\hat{\beta}_i \neq 0$

et la statistique de test se simplifie donc lorsque l'on suppose $H_0$ vraie :

$$ \boxed{t_{\hat{\beta}_i } = \frac{\hat{\beta}_i}{\hat{\sigma}_{\hat{\beta}_i }} \sim t(T-k-1) }$$

où $\hat{\sigma}_{\hat{\beta}_i } = \hat{\sigma}_{\epsilon} \sqrt{a_{i+1,i+1}}$, est l'écart type estimé de $\hat{\beta}_i $. La régle de décision devient alors :


- Si $\vert t_{\hat{\beta}_i} \vert \leq t_{p/2}$, on ne rejette pas l'hypothèse nulle au seuil statistique de $100p\%$ et on considèe que $\hat{\beta}_i = 0$. <b>La variable $X_i$ est dite non statistiquement significative</b>, elle n'influence pas $Y$.
- Si $\vert t_{\hat{\beta}_i} \vert > t_{p/2}$, on rejette l'hypothèse nulle au seuil statistique de $100p\%$ et on considèe que $\hat{\beta}_i \neq 0$. On dit alors que <b>la variable est statistiquement significative</b>.

## Tests sur plusieurs coefficients : test de Fisher <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p4.3'></a>

Le test de Student nous permet de tester la significativité statistique d'une variable seule. Nous procédons maintenant à la description du test de Fisher, qui permet de tester la significativité de plusieurs coefficients simultanément. 

- Idée générale du test : poser une contrainte sur plusieurs coefficients en même temps,

$$R \beta = r $$

où $R$ est une matrice de plein rang et de taille $(q,k+1)$ avec $q$ le nombre de contraintes, et $r$ le vecteur des contriantes de taille $(q,1)$.

- Exemple 1 : pour tester si le coefficeint $\beta_1$ est égal à $0$,

$$R = \begin{pmatrix} 0 & 1 & 0 & ... & 0 \end{pmatrix}$$

$$\beta = 
\begin{pmatrix}
\alpha \\
\beta_1 \\
\beta_2 \\
\vdots  \\
\beta_k \\
\end{pmatrix}
$$

$$r = 0$$

et on a : $R \beta = r $, ce qui donne $\beta_1 = 0$

<br>

- Exemple 2 : si l'on veut tester l'égalité entre $\beta_1$ et $\beta_2$ ;

$$R = \begin{pmatrix} 0 & 1 & -1 & ... & 0 \end{pmatrix}$$


$$r = 0$$

et on a : $R \beta = r $, ce qui donne $\beta_1 - \beta_2 = 0$, c a d $\beta_1 = \beta_2$.

<br>

- Ainsi, nous allons nous servir de ce test pour tester la significativité statistique globale des coefficients :

$$R = 
\begin{pmatrix}
0 & 1 & 0 & ... & 0\\
0  & 0 & 1 & ... & 0\\
\vdots & \vdots & \vdots & ... & \vdots\\
0 & 0 & 0 & ... & 1 \\
\end{pmatrix}
$$

$$r = 
\begin{pmatrix}
0 \\
0 \\
0 \\
0 \\
0 \\
\end{pmatrix}
$$

de sorte que l'on teste au final : $\beta_1 = \beta_2 = ... = \beta_k = 0$. On teste l'hypothèse nulle "aucun coefficient n'est significatif et n'influence la variable expliquée".

<br>



Afin de mener le test, nous procédons ainsi : nous allons comparer 
- les résidus du modèle non-contriant($SCR_{nc}$) (estimation classique de $Y$ sur l'ensemble des variables explicatives) 
- ceux du modèle contraint ($SCR_{c}$) (régression de $Y$ sur la constante dans le dernier cas). 
- Cela donne une statistique de test : 

$$\boxed{ F = \frac{SCR_{nc} - SCR_c}{SCR_{nc}/(T-k-1)} \sim F(q,T-k-1)}$$ 

Nous verrons à la section suivante que, dans le cadre du test de significativité globale, <b>et uniquement dans ce cas là</b>, nous pouvons réécrire cette statitique plus simplement : 

$$\boxed{ F = \frac{R^2/k}{(1-R^2)/(T-k-1)} \sim F(k,T-k-1)}$$

# Analyse de la variance <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p5'></a>


## Equation de l'analyse de la variance <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p5.1'></a>

En partant de l'équation du modèle : 

$$Y = X\beta + \epsilon $$

nous avons obtenus, après estimation, une expréssion pour le résidu :

$$e = y - x\hat{\beta}$$

où l'on note $X$ et $Y$ en minuscule pour signifier qu'elles sont centrées (on leur retire leur moyenne). Nous pouvons en déduire une nouvelle expression d'écriture de la somme des carrés des résidus : 

$$e'e = (y - x\hat{\beta})'(y - x\hat{\beta})\\
= y'y - 2\hat{\beta} x'y + \hat{\beta}'x'x\hat{\beta}$$

et après réarangement, nous obtenons l'équation d'analyse de la variance :


$$\boxed{ y'y = \hat{\beta}'x'y + e'e  }$$

où :
- la somme des carrés totale (SCT) est donnée par : $y'y = \sum_t y_t^2 = \sum_t (Y_t - \bar{Y})^2$
- la somme des carrés expliquée (SCE) est donnée par : $\hat{\beta}'x'y$
- la somme des carrés des résidus (SCR) est donnée par : $e'e = \sum_t e_t^2 $

On peut retrouver la même équation pour le cas de variables non centrées. 

</span>

<span style='line-height:1.5;font-size: 17px;font-family:"Times New Roman"'>

## Coefficient de détermination et coefficient de détermination corrigé <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p5.2'></a>

<b>Le coefficeint de détermination ($R^2$)</b> : mesure la part de la variance totale expliquée par les variables explicatives et permet de juger de la qualité de l'ajustement du modèle.

$$ \boxed{R^2 = \frac{SCE}{SCT} = 1 - \frac{SCR}{SCT}}$$


où $0 \leq R^2 \leq 1$.

<br>

<b>Le coefficeint de détermination corrigé/ajusté ($\bar{R}^2$)</b> : 
- le coefficient de détermination classique augmente à avec le nombre de variables explicatives intégrées dans le modèle (puisque l'introduction de nouvelles variables ne peut pas faire diminuer la SCR).
- on peut donc le corriger de ce biais :

$$\bar{R}^2 = 1 - \frac{T-1}{T-k-1}(1-R^2)$$

#  Prévision <font color='blue' size="1.5"><i>[Plan du cours](#plan)</i></font><a id='p7'></a>


- Estimation d'un modèle : 
    - permet d'expliquer ce qui est observé : pouvoir explicatif (en échantillon)
    - permet d'émettre des prévision : pouvoir prédictif du modèle (hors échantillon)
    
<br>

On se place dans une situation où le modèle suivant a été estimé :

$$Y_t = \alpha + \beta_1 X_{1,t} + \beta_2 X_{2,t} + ... + \beta_k X_{k,t} + \epsilon_t$$

Ainsi, nous avons une valeur estimée de $Y$ (la variable que l'on cherche à expliquer) qui est donnée par :

$$\hat{Y}_t = \hat{\alpha} + \hat{\beta}_1 X_{1,t} + \hat{\beta}_2 X_{2,t} + ... + \hat{\beta}_k X_{k,t}$$
    
La prévision consiste donc à estimer la valeur de $Y$, mais pour un horizon $h$, c a d $\hat{Y}_{t+h}$. L'intervalle de prévision est donné par :

$$\hat{Y}_{t+h} \pm t_{p/2} \sqrt{R(X'X)^{-1}R' +1}$$

où $R$ est la matrice des variables explicative pour la date $t+h$.


<br>
    
    
- Il existe plusieurs cas de prévision : 
    - si les valeurs des variables explicatives sont connues en $t+h$ (c a d $R$) alors on peut calculer la prévision et sont intervalle (au seuil critique de $p$)
    - sinon, il est possible de faire de l'hors échantillonage pour se faire une idée du pouvoir prédictif du modèle

<br>

<b>Hors échantillonage</b> 

Une manière de s'intéresser au pouvoir prédictif du modèle est de ne pas l'estimer sur tout l'échantillon dont on dispose : 
- on estime le modèle sur un partie de l'échantillon
- puis on regarde l'erreur de prévision sur la partie restant de l'échantillon

C'est une technique particulièrement utilisée dans le paramètrage des modèles d'apprentissage statistique. Cette méthode permet en outre de s'intéresser au caractère intertemporel d'une estimation ("est-elle bonne que dans le cadre des observations que l'on a, ou a-t-elle un pouvoir explicatif au delà de l'échantillon?").

Le renforcement des grandes bases de données en finance permet de plus en plus de type d'approche. 

</span>
