# Financial Analysis Project : A financial analysis of ALTAREA

## Importing packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

## Importing the dataset

```python
path = "C:\\Users\\issak\\Documents\\M1\\Analyse financi√®re\\PROJET_ANALYSE_FINANCIERE\\"

# Importing daily data from CAC SMALL index

df_CAC = pd.read_csv(path + "CAC_SMALL_historical_price.csv", sep=";")

# Importing daily data from ALTAREA

df_ALTA = pd.read_csv(path + "ALTAREA_historical_price.csv", sep=";")

# First visualisation of the dataframe 


print("CAC SMALL Index", df_CAC.head())
print('\n')
print("ALTAREA", df_ALTA.head())
```

## Creating a new data frame

```python
# Creating a new data frame whose only columns are 'Date' and 'Close'.

df_CAC = df_CAC[['Date','Close']]

df_ALTA = df_ALTA[['Date','Close']]
# visualisation of the dataframe 

print("Close price of the CAC SMALL",df_CAC.head())
print('\n')
print("Close price of the ALTAREA",df_ALTA.head())
```

```python
# Sorting the data by date from the oldest to the newest

df_CAC['Date'] =  pd.to_datetime(df_CAC['Date'], dayfirst=True,errors='coerce')
df_CAC = df_CAC.sort_values(by=['Date'],axis=0, ascending=True)

df_ALTA['Date'] =  pd.to_datetime(df_ALTA['Date'], dayfirst=True,errors='coerce')
df_ALTA = df_ALTA.sort_values(by=['Date'],axis=0, ascending=True)

df_CAC = df_CAC.reset_index(drop=True).copy()
df_ALTA = df_ALTA.reset_index(drop=True).copy()

# visualisation of the dataframe 

print("Close price of the CAC SMALL",df_CAC.head(10))
print('\n')
print("Close price of the ALTAREA",df_ALTA.head(10))
```

## Plotting the data

```python
# plotting the data

plt.figure(figsize=(20,10))
plt.plot(df_CAC['Date'], df_CAC['Close'], label='Close Price', color='red')
plt.title("CAC SMALL index")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=100))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(nbins=20))
plt.tight_layout()
plt.show()

plt.figure(figsize=(20,10))
plt.plot(df_ALTA['Date'], df_ALTA['Close'], label='Close Price', color='darkviolet')
plt.title("ALTAREA")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=100))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(nbins=20))
plt.tight_layout()
plt.show()
```

## Creating a new data frame for 2023

```python
# Creating a new dataframe with only the close price of the CAC SMALL index and ALTAREA for the year 2023

df_CAC_2023 = df_CAC[df_CAC['Date'].dt.year == 2023]
print("Close price of the CAC SMALL for 2023"+'\n \n', df_CAC_2023.head(10))
print('\n')
df_ALTA_2023 = df_ALTA[df_ALTA['Date'].dt.year == 2023]
print("Close price of ALTAREA for 2023"+'\n \n', df_ALTA_2023.head(10))

# plotting the data

plt.figure(figsize=(20,10))
plt.plot(df_CAC_2023['Date'], df_CAC_2023['Close'], label='Close Price', color='red')

plt.title("CAC SMALL index in 2023")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=100))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(nbins=20))
plt.tight_layout()
plt.show()

plt.figure(figsize=(20,10))
plt.plot(df_ALTA_2023['Date'], df_ALTA_2023['Close'] , label='Close Price', color='darkviolet')
plt.title("CAC SMALL index in 2023")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=100))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(nbins=20))
plt.tight_layout()
plt.show()
```

## Computing the annual return 

```python
# Computing the annual return for the CAC SMALL index and ALTAREA in 2023

df_CAC_2023.shape[0]
df_ALTA_2023.shape[0]

return_CAC_2023 = (df_CAC_2023['Close'][240]/df_CAC_2023['Close'][0]-1)*100
return_ALTA_2023 = (df_ALTA_2023['Close'][240]/df_ALTA_2023['Close'][0]-1)*100

if return_CAC_2023 > 0 :
    print('The annual return of the CAC SMALL in 2023 is : +', return_CAC_2023,'%')
else :
    print('The annual return of the CAC SMALL in 2023 is :', return_CAC_2023,'%')
if return_ALTA_2023 > 0 :
    print('The annual return of ALTAREA in 2023 is : +', return_ALTA_2023,'%')
else :
    print('The annual return of ALTAREA in 2023 is :', return_ALTA_2023,'%')        
```

```python
# Computing the daily return for the CAC SMALL index and ALTAREA in 2023

df_CAC_2023 = df_CAC_2023[df_CAC_2023['Date'].dt.year == 2023].copy()
df_CAC_2023['CAC_SMALL_return'] = np.log(df_CAC_2023['Close']/df_CAC_2023['Close'].shift(1))
print(df_CAC_2023.head(10))
print('\n')
df_ALTA_2023 = df_ALTA_2023[df_ALTA_2023['Date'].dt.year == 2023].copy()
df_ALTA_2023['ALTA_return'] = np.log(df_ALTA_2023['Close']/df_ALTA_2023['Close'].shift(1))
print(df_ALTA_2023.head(10))
```

## Computing the descriptive statistics

```python
# Computing the mean and the standard deviation for the CAC SMALL index in 2023
#df_CAC_2023.dropna()
#df_ALTA_2023.dropna()
print(df_CAC_2023[['CAC_SMALL_return']].describe())
print('\n')
print(df_ALTA_2023[['ALTA_return']].describe())
```

### $\text{Coeffiecient de variation} = valeur abolue \frac{\sigma_X}{\bar{X}}$

On peut chercher √† calculer un intervalle de confiance

```python
# Computing the variance, the skewness and the kurtosis of the closing price for the CAC SMALL index and ALTAREA in 2023.

var_CAC_2023 = df_CAC_2023['CAC_SMALL_return'].var()
skew_CAC_2023 = df_CAC_2023['CAC_SMALL_return'].skew()
kurt_CAC_2023 = df_CAC_2023['CAC_SMALL_return'].kurt()
print('The variance for the close price of the CAC SMALL index in 2023 is', var_CAC_2023)
print('The skewness for the close price of the CAC SMALL index in 2023 is', skew_CAC_2023)
print('The kurtosis for the close price of the CAC SMALL index in 2023 is', kurt_CAC_2023)

print('\n')

var_ALTA_2023 = df_ALTA_2023['ALTA_return'].var()
skew_ALTA_2023 = df_ALTA_2023['ALTA_return'].skew()
kurt_ALTA_2023 = df_ALTA_2023['ALTA_return'].kurt()
print('The variance for the close price of ALTAREA in 2023 is', var_ALTA_2023)
print('The skewness for the close price of ALTAREA in 2023 is', skew_ALTA_2023)
print('The kurtosis for the close price of ALTAREA in 2023 is', kurt_ALTA_2023)
```

## Plotting the data of the return for the CAC SMALL index and ALTAREA in the 2023

```python
### plotting the data of the return for the CAC SMALL index and ALTAREA in the 2023

plt.figure(figsize=(20,10))
plt.plot(df_CAC_2023['Date'], df_CAC_2023['CAC_SMALL_return'] , label='Daily return of the CAC SMALL index', color='red')
plt.plot(df_ALTA_2023['Date'], df_ALTA_2023['ALTA_return'] , label='Daily return of ALTAREA', color='darkviolet')
plt.title("Daily return of the CAC SMALL and ALTARE in 2023")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=100))
plt.tight_layout()
plt.show()
```
