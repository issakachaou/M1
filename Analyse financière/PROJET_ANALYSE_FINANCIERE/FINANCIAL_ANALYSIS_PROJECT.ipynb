## Computing the annual return 

```python
# Computing the annual return for the CAC SMALL index and ALTAREA in 2023

df_CAC_2023.shape[0]
df_ALTA_2023.shape[0]

return_CAC_2023 = (df_CAC_2023['Close'][172]/df_CAC_2023['Close'][0]-1)*100
return_ALTA_2023 = (df_ALTA_2023['Close'][172]/df_ALTA_2023['Close'][0]-1)*100

if return_CAC_2023 > 0 :
    print('The annual return of the CAC SMALL in 2023 is : +', return_CAC_2023,'%')
else :
    print('The annual return of the CAC SMALL in 2023 is :', return_CAC_2023,'%')
if return_ALTA_2023 > 0 :
    print('The annual return of ALTAREA in 2023 is : +', return_ALTA_2023,'%')
else :
    print('The annual return of ALTAREA in 2023 is :', return_ALTA_2023,'%')        
```

```python
# Computing the daily return for the CAC SMALL index and ALTAREA in 2023

df_CAC_2023 = df_CAC_2023[df_CAC_2023['Date'].dt.year == 2023].copy()
df_CAC_2023['CAC_SMALL_return'] = np.log(df_CAC_2023['Close']/df_CAC_2023['Close'].shift(1))*100
print(df_CAC_2023.head(10))
print('\n')
df_ALTA_2023 = df_ALTA_2023[df_ALTA_2023['Date'].dt.year == 2023].copy()
df_ALTA_2023['ALTA_return'] = np.log(df_ALTA_2023['Close']/df_ALTA_2023['Close'].shift(1))*100
print(df_ALTA_2023.head(10))
```

## Computing the descriptive statistics

```python
# Computing the mean and the standard deviation for the CAC SMALL index in 2023
#df_CAC_2023.dropna()
#df_ALTA_2023.dropna()
print(df_CAC_2023[['CAC_SMALL_return']].describe())
print('\n')
print(df_ALTA_2023[['ALTA_return']].describe())
```

### $\text{Coeffiecient de variation} = valeur abolue \frac{\sigma_X}{\bar{X}}$

On peut chercher √† calculer un intervalle de confiance

```python
# Computing the variance, the skewness and the kurtosis of the closing price for the CAC SMALL index and ALTAREA in 2023.

var_CAC_2023 = df_CAC_2023['CAC_SMALL_return'].var()
skew_CAC_2023 = df_CAC_2023['CAC_SMALL_return'].skew()
kurt_CAC_2023 = df_CAC_2023['CAC_SMALL_return'].kurt()
print('The variance for the close price of the CAC SMALL index in 2023 is', var_CAC_2023)
print('The skewness for the close price of the CAC SMALL index in 2023 is', skew_CAC_2023)
print('The kurtosis for the close price of the CAC SMALL index in 2023 is', kurt_CAC_2023)

print('\n')

var_ALTA_2023 = df_ALTA_2023['ALTA_return'].var()
skew_ALTA_2023 = df_ALTA_2023['ALTA_return'].skew()
kurt_ALTA_2023 = df_ALTA_2023['ALTA_return'].kurt()
print('The variance for the close price of ALTAREA in 2023 is', var_ALTA_2023)
print('The skewness for the close price of ALTAREA in 2023 is', skew_ALTA_2023)
print('The kurtosis for the close price of ALTAREA in 2023 is', kurt_ALTA_2023)
```

```python
df_CAC_2023 = df_CAC_2023.dropna()
df_ALTA_2023 = df_ALTA_2023.dropna()

# Calcul de la droite de tendance  
coeffs = np.polyfit(df_CAC_2023['CAC_SMALL_return'], df_ALTA_2023['ALTA_return'], 1)  
p = np.poly1d(coeffs)  

# Trac√© du graphique  
plt.figure(figsize=(20,10))  
plt.title("Relation entre les rendements journaliers d'ALTAREA et du CAC SMALL en 2023")  
plt.ylabel("Rendement d'ALTAREA")  
plt.xlabel("Rendement du CAC SMALL")  
plt.scatter(df_CAC_2023['CAC_SMALL_return'], df_ALTA_2023['ALTA_return'], color='blue', alpha=0.7, label="Rendements journaliers")  

# Trac√© de la droite de tendance  
x_vals = np.linspace(df_CAC_2023['CAC_SMALL_return'].min(), df_CAC_2023['CAC_SMALL_return'].max(), 100)  
plt.plot(x_vals, p(x_vals), "r", label="Droite de tendance")  

plt.legend()  
plt.grid(True)  
plt.show()
```

### Computing the market model : $\text{Return}_t^{ALTAREA} = \alpha + \beta \cdot \text{Return}_t^{CAC}+\epsilon_t$

$ f(\beta,\sigma^2_{CAC})$

Mesure du choc sp√©cifique = $1-R^2$

```python
X = df_CAC_2023['CAC_SMALL_return']
Y = df_ALTA_2023['ALTA_return']
X = sm.add_constant(X)
model = sm.OLS(Y,X)
resultat = model.fit()
print(resultat.summary())
```
